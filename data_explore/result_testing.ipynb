```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("../runs_11n/detect/train/weights/best.pt")
results = model.predict(source="../dataset/roadpoles_preprocessed/valid/images/frame_000000.PNG")
plot = results[0].plot()
```

```python
cv2.imshow("Snow Pole Detection", plot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
