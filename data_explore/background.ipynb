```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../dataset/TabellSVV.csv", sep=";")
df = df.sort_values("År")
```

```python
plt.figure(figsize=(14, 7))
plt.bar(df["År"], df["Lettere skadd"], label="Lettere skadd", color="green")
plt.bar(df["År"], df["Hardt skadd"], bottom=df["Lettere skadd"], label="Hardt skadd", color="orange")
plt.bar(df["År"], df["Drept"], bottom=df["Lettere skadd"] + df["Hardt skadd"], label="Drept", color="red")

plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.title("Number of snow related accidents in Norway (1990 - 2024)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("../dataset/Drepte og hardt skadde.csv", sep=';',header=None,skiprows=2,encoding='utf-8-sig')
df_gender = df[df[0].isin(['I alt', 'Menn', 'Kvinner'])].copy()
```

```python
df_gender[0] = df_gender[0].replace({
    'I alt': 'Total',
    'Menn': 'Men',
    'Kvinner': 'Women'
})
```

```python
df_gender.columns = ['Kategori', 'Drept_2025', 'Drept_2024', 'Hardt_skadd_2025', 'Drept_snitt_5', 'Hardt_skadd_snitt_5']
```

```python
plt.figure(figsize=(12, 6))

plt.bar(df_gender['Kategori'], df_gender['Hardt_skadd_2025'], label='Hardt skadd', color='orange')
plt.bar(df_gender['Kategori'], df_gender['Drept_2025'], bottom=df_gender['Hardt_skadd_2025'], label='Drept', color='red')

plt.xlabel('Gender')
plt.ylabel('Number of Accidents')
plt.title('Number of snow related accidents in Norway (1990 - 2024)')
plt.legend()
plt.show()
```
